```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
profile_df = pd.read_csv('profile.csv')
profile_df
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='operation', y='execution time', data=profile_df.query("operation=='evalAdd'"), hue='num_threads')
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='operation', y='execution time', data=profile_df.query("num_threads in [1, 2, 4, 8, 16]").query("operation in ['Generating', 'Encrypting', 'bootstrap']"), hue='num_threads')
```
